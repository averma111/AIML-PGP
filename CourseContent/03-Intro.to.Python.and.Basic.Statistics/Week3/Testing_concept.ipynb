```python
import scipy.stats as zscore
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

<H1> Compute Outliers

```python
test_outliers = [-3,1, 2, 2, 3, 3, 4, 5, 5, 9,9.6]

outliers_quartiles = np.percentile(test_outliers,[25,50,75],interpolation='midpoint')
print(outliers_quartiles)

interquartile_range = outliers_quartiles[2]-outliers_quartiles[0]
print(interquartile_range)

minimum_value = outliers_quartiles[0]-1.5 * interquartile_range
maximum_value = outliers_quartiles[2]+1.5 * interquartile_range


print('Minimum of the range is ', minimum_value,' & maximum is',maximum_value)

plt.boxplot(test_outliers,vert=False)
```

<H1> Positively Skewed Distribution 

```python
x,y = [],[]

for i in range(100):
    x.append(i)
    
    y_new = i*i*i+2*i*i - 2*i +random.randint(1,10000)
    
    y.append(y_new)
    


data_exponential = pd.DataFrame(columns= ['X','Y'])

data_exponential.X = x
data_exponential.Y = y

plt.figure(figsize=(8,8))
sns.distplot(data_exponential.Y,)

exp_avg = np.average(data_exponential.Y)


exp_quartile = np.percentile(data_exponential.Y,[25,50,75], interpolation= 'midpoint')

print( 'The Mean is ', exp_avg , '& the Median is ', exp_quartile[1])



```

```python
log_exponetial = np.log(data_exponential['Y'])
sns.distplot(log_exponetial)

log_exp_avg = np.average(log_exponetial)


log_exp_quartile = np.percentile(log_exponetial,[25,50,75], interpolation= 'midpoint')

print( 'The Mean is ', log_exp_avg , '& the Median is ', log_exp_quartile[1])
```
